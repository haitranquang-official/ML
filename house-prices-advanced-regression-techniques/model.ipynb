```python
import pandas as pd
import numpy as np
import json
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
```

```python
f = pd.read_csv("train.csv")
f.head()
```

```python
datas = pd.DataFrame()
datas = f
datas = shuffle(datas)
datas.head()
```

```python
datas.info()
```

```python
datas.describe()
```

```python
import seaborn as sns
sns.distplot(datas['SalePrice'])
```

```python
correlation_matrix = datas.corr()
sns.heatmap(correlation_matrix, vmax = 1, square = True)
datas = datas.drop(columns=['Id'])
```

```python
numerical_data = datas.select_dtypes(include = ['float64', 'int64'])
numerical_data.head()
```

```python
numerical_data.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);
```

```python
for i in range(0, len(numerical_data.columns), 5):
    sns.pairplot(data=numerical_data,
                x_vars=numerical_data.columns[i:i+5],
                y_vars=['SalePrice'])
```

```python
non_numerical_data = datas.select_dtypes(include = 'O')
print(non_numerical_data.columns.tolist())
```

```python
fig, axes = plt.subplots(round(len(non_numerical_data.columns) / 3), 3, figsize=(12, 30))

for i, ax in enumerate(fig.axes):
    if i < len(non_numerical_data.columns):
        ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=45)
        sns.countplot(x=non_numerical_data.columns[i], alpha=0.7, data=non_numerical_data, ax=ax)

fig.tight_layout()
```
