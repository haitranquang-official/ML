```python
import pandas as pd
import numpy as np
import json
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
f = pd.read_csv("train.csv")
f.head()
```

```python
datas = pd.DataFrame()
datas = f
datas = shuffle(datas)
datas.head()
```

```python
datas.info()
```

```python
datas.describe()
```

```python
total = datas.isnull().sum().sort_values(ascending=False)
percent = (datas.isnull().sum()/datas.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
colums_fillable = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'LotFrontage']

for column in colums_fillable:
    datas[column].fillna('None', inplace=True)

# datas.fillna(datas.mean(), inplace=True)
datas = datas.dropna()
datas.shape

```

```python
import seaborn as sns
sns.distplot(datas['SalePrice'])
```

```python
datas = datas.drop(columns=['Id'])
```

```python
correlation_matrix = datas.corr()
sns.heatmap(correlation_matrix, vmax = 1, square = True)
```

```python
numerical_data = datas.select_dtypes(include = ['float64', 'int64'])
numerical_data.head()
```

```python
corr = datas.corr()
corr = corr.abs()
num_columns = len(numerical_data.columns)
high_corr = corr.nlargest(num_columns, 'SalePrice')['SalePrice']
numerical_columns_to_drop = list(high_corr[high_corr.values < 0.2].index) 
```

```python
numerical_columns_to_drop
```

```python
datas = datas.drop(columns=numerical_columns_to_drop)
numerical_data = numerical_data.drop(columns=numerical_columns_to_drop)
```

```python
datas.shape
```

```python
numerical_data.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);
```

```python
for i in range(0, len(numerical_data.columns), 5):
    sns.pairplot(data=numerical_data,
                x_vars=numerical_data.columns[i:i+5],
                y_vars=['SalePrice'])
```

```python
non_numerical_data = datas.select_dtypes(include = 'O')
print(non_numerical_data.columns.tolist())
```

```python
fig, axes = plt.subplots(round(len(non_numerical_data.columns) / 3), 3, figsize=(12, 30))

for i, ax in enumerate(fig.axes):
    if i < len(non_numerical_data.columns):
        ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=45)
        sns.countplot(x=non_numerical_data.columns[i], alpha=0.7, data=non_numerical_data, ax=ax)

fig.tight_layout()
```

```python
categorical_columns_to_drop = ['Condition2', 'Street', 'Alley', 'Utilities', 'RoofMatl', 'Functional',
'PoolQC', 'Heating', 'MiscFeature']
non_numerical_data = non_numerical_data.drop(columns=categorical_columns_to_drop)
datas = datas.drop(columns = categorical_columns_to_drop)
```

```python
fig, axes = plt.subplots(round(len(non_numerical_data.columns) / 3), 3, figsize=(12, 30))

for i, ax in enumerate(fig.axes):
    if i < len(non_numerical_data.columns):
        ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=45)
        sns.countplot(x=non_numerical_data.columns[i], alpha=0.7, data=non_numerical_data, ax=ax)

fig.tight_layout()
```

```python
for category in non_numerical_data.columns:
    sns.violinplot(x=category, y='SalePrice', data=datas)
    plt.show()
```
